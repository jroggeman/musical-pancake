```python
from stream_json import stream_json
```

```python
e = stream_json(filename='../data/smaller_reviews.json')
```

```python
import itertools

examples = list(itertools.islice(e, 50))

examples, votes = zip(*[(e['text'], e['votes']['useful'] > 0) for e in examples])
```

```python
votes
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
```

```python
counts = vectorizer.fit_transform(examples).toarray()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

counts_new = SelectKBest(chi2, k=20).fit_transform(counts, votes)
```

```python
counts_new
```

```python
from pybrain.tools.shortcuts import buildNetwork

net = buildNetwork(20, 10, 1)
```

```python
from pybrain.datasets import SupervisedDataSet
```

```python
ds = SupervisedDataSet(20, 1)
```

```python
for c, v in zip(counts_new, votes):
    ds.addSample(c, v)
```

```python
from pybrain.supervised.trainers import BackpropTrainer
```

```python
trainer = BackpropTrainer(net, ds)
```

```python
trainer.trainUntilConvergence()
```

```python
net.activate(examples[0])
```

```python
num_correct = 0

for index, c in enumerate(counts_new):
    g = net.activate(c)
    print("Guess: " + str(g))
    print("Actual: " + str(votes[index]))
    if (g > 0.5) == votes[index]:
        num_correct += 1
        
print("Accuracy: " + str(float(num_correct) / len(counts_new)))
```
